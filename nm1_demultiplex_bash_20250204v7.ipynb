```python
from demuxalot import Demultiplexer, BarcodeHandler, ProbabilisticGenotypes, count_snps
import pandas as pd
import re
```

```python

genotypes = ProbabilisticGenotypes(genotype_names=['1-1','1-3','1-5','1-6','1-7','1-8','2-1','2-2'
                                                   ])
genotypes.add_vcf('/data/scrna/scrna_snubh_batch20250106_counts/vcf_merged/GEX1_jeong_final_chr_fixed.vcf')

print(f'Loaded genotypes: {genotypes}')
```

```python
barcode_handler = BarcodeHandler.from_file('/data/scrna/scrna_snubh_batch20250106_counts/GEX1_jeong/outs/raw_feature_bc_matrix/barcodes.tsv.gz')
print(f'Loaded barcodes: {barcode_handler}')
```

```python

snps = count_snps(
    bamfile_location='/data/scrna/scrna_snubh_batch20250106_counts/GEX1_jeong/outs/possorted_genome_bam.bam',
    chromosome2positions=genotypes.get_chromosome2positions(),
    barcode_handler=barcode_handler,
)

print('Collected SNPs: ')
for chromosome, snps_in_chromosome in snps.items():
    print(f'Chromosome {chromosome}, {snps_in_chromosome.n_snp_calls} calls in {snps_in_chromosome.n_molecules} mols')

```

```python
# returns two dataframes with likelihoods and posterior probabilities
log_likelihoods, posterior_probabilities = Demultiplexer.learn_genotypes(
    snps,
    genotypes=genotypes,
    barcode_handler=barcode_handler,
    doublet_prior=0.25,
)

print('Result:')
print(posterior_probabilities.round(3))
```

```python
row_series = posterior_probabilities.iloc[2]

# Get the ten columns with the highest numeric values
top10 = row_series.nlargest(10)

# Print the Series directly
print("Top 10 columns with highest values:")
print(top10)

print(posterior_probabilities.iloc[3].sum())
```

```python
def classify_row(row):
    max_value = row.max()
    best_col = row.idxmax()
    
    #### Here we set the threshold #### 
    if max_value < 0.5:
        return "ambivalent"
    else:
        if re.fullmatch(r'\d+-\d+\+\d+-\d+', best_col):
            return "doublet"
        elif re.fullmatch(r'\d+-\d+', best_col):
            return "singlet"
        else:
            return "ambivalent"  
        
posterior_probabilities['call'] = posterior_probabilities.apply(classify_row, axis=1)

```

```python
print(posterior_probabilities['call'])

counts = posterior_probabilities['call'].value_counts()
print("Counts for each call:")
print(counts)

# Or use .describe() for a summary of the categorical data:
summary = posterior_probabilities['call'].describe()
print("\nSummary statistics for 'call':")
print(summary)
```
